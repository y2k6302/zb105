## 把切割出來的字串分別存取在兩個變數內

```python
with open('bid_list.txt', 'r') as f:
    for url in f.readlines():
        rearstr = url.strip().split('pkAtmMain=')
        
#pkatmmain, caseno 中間用逗號隔開，會把list的元素分別存在兩個變數內
        pkatmmain, caseno = rearstr[1].split('&tenderCaseNo=')
        filename = 'gov/{}_{}.txt'.format(pkatmmain, caseno)
        #print filename
        with open(filename, 'w') as f2:
            res = requests.get(url.strip())
            f2.write(res.text.encode('utf-8'))
```

## Time 睡覺模式

```python
import time
with open('bid_list.txt', 'r') as f:
    for url in f.readlines():
        rearstr = url.strip().split('pkAtmMain=')
        pkatmmain, caseno = rearstr[1].split('&tenderCaseNo=')
        filename = 'gov/{}_{}.txt'.format(pkatmmain, caseno)
        #print filename
        with open(filename, 'w') as f2:
            res = requests.get(url.strip())
            f2.write(res.text.encode('utf-8'))
        time.sleep(1)
```

```python
#timedelta 可以加天數(只限日、月、周、不能用年)
from datetime import date,datetime, timedelta
currentdt=datetime.today()
print currentdt + timedelta(days= 1)
```
