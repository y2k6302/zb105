## GET 模擬向伺服器送出請求

```python
import requests
url = 'https://www.technologyreview.com/'
res = requests.get(url) #get是去模仿使用者用瀏覽器向伺服器請求資訊的動作
```

## 將網站資訊排整齊，並獲得鎖定的資訊(ex.標題)

```python
#BeautifulSoup 放在bs4 裡面的套件裡面
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text)  #請美麗幫忙排整齊，準備抓標題
h2 = soup.select('.hp-daily__tz__title')    #.class選擇器，要抓此class的東西，select()是soup的方法
for h in h2 :
    print h.text
```

## POST 模擬向伺服器送出請求

```python
import requests
url = 'http://www.thsrc.com.tw/tw/TimeTable/SearchResult'
playload = {
    'StartStation':'977abb69-413a-4ccf-a109-0272c24fd490',
    'EndStation':'f2519629-5973-4d08-913b-479cce78a356',
    'SearchDate':'2016/02/22',
    'SearchTime':'11:00',
    'SearchWay':'DepartureInMandarin',
    'RestTime':'',
    'EarlyOrLater':''
}
res = requests.post(url, data = playload)  #post 模擬傳一個信封給伺服器，傳送內容就是 playload
soup = BeautifulSoup(res.text)
trs = soup.select('tr')   #trs是一個list
for tr in trs :
    print tr.text.strip()  #strip()是把前後的空白刪掉
```

## 補充:當網站內文出現亂碼時

```python
#利用 encoding 把網站的編碼統一指定為"該網頁設定的編碼"
url = 'https://www.thsrc.com.tw/index.html?force=1'
res = requests.get(url)
res.encoding = 'utf-8'   #把request整個東西重新指定編碼
print res.text
```

## 如何延續Session

```python
front_page = "http://web.pcc.gov.tw/tps/pss/tender.do?searchMode=common&searchType=basic"
playload = {
'method':'search',
'searchMethod':'true',
'tenderUpdate':'',
'searchTarget':'',
'orgName':'',
'orgId':'',
'hid_1':'1',
'tenderName':'',
'tenderId':'',
'tenderType':'tenderDeclaration',
'tenderWay':'1,2,3,4,5,6,7,10,12',
'tenderDateRadio':'on',
'tenderStartDateStr':'105/02/12',
'tenderEndDateStr':'105/02/18',
'tenderStartDate':'105/02/12',
'tenderEndDate':'105/02/18',
'isSpdt':'N',
'proctrgCate':'3',
'radProctrgCate':'3',
'btnQuery':'查詢',
'hadUpdated':''
}
rs = requests.session()     #rs為可以存session的物件
rs.post(front_page,data=playload)   #利用rs跑一次流程，把Session都存起來，之後只要用 rs就可以get任何東西，並帶有前面的session
url = 'http://web.pcc.gov.tw/tps/pss/tender.do?searchMode=common&searchType=basic&method=search&isSpdt=&pageIndex=3'
res = rs.get(url)     
print res.text
```

## 寫入 file

```python
dib_url = 'http://web.pcc.gov.tw/tps/tpam/main/tps/tpam/tpam_tender_detail.do?searchMode=common&scope=F&primaryKey=51813202'
res_bid = requests.get(dib_url)
soup = BeautifulSoup(res_bid.text)
to_file = soup.select('#print_area')[0]
file_id = dib_url.split('primaryKey=')[1]   #設計檔名，以primaryKey=做為切割點
f = open('/Users/BIG DATA/Desktop/{}.text'.format(file_id),'w')
f.write(to_file.prettify().encode('utf-8')) #prettify()先把檔案轉unicode，再用encode('utf-8')編碼
f.close()
```
