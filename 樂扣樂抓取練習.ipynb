```python
import requests
res = requests.get('https://www.flyingv.cc/type/online')
```

```python
from bs4 import BeautifulSoup as bs
soup = bs(res.text)

divs = soup.select('.portfolio-item-wrapper')  
for div in divs:
    title = div.select('.portfolio-content h5')[0].text
    person = div.select('.unit-owner')[0].text.strip()
    price = div.select('.unit-money.flyingred')[0].text
    time = div.select('.unit-time')[0].text.strip()
    print title,person,price,time
```
